## 方式一：直接传入sql 语句，从clickhouse 中读取数据

from plutus.utils.database.ck_control import ClickHouse
ck_control = ClickHouse()

```python
all_1d_df = ck_control.read_df_by_sql_str(
    "select * from commercial_cn_stock_md.all_1d where trading_date > '2021-04-19 00:00:00.000' order by trading_date desc "
)
all_1d_df.head()
```

## 方式二：传入参数，从数据库中读取数据

```python
from plutus.utils.database.database_control import DatabaseControl

db_control = DatabaseControl()
```

```python
all_1d = db_control.read_clickhouse(
    db="commercial_cn_stock_md",
    tb="all_1d",
    filters=[("trading_date", ">", "'2021-04-19 00:00:00.000'")],
)
all_1d.head()
```
